# OBSERVACIONES

1.	Eliminación del Prefijo b':
o	Los datos de titulares de noticias originalmente contenían un prefijo b', lo que indicaba que las cadenas habían sido almacenadas como bytes. Este prefijo fue removido para facilitar el análisis y procesamiento posterior del texto. Este paso es esencial para trabajar con los titulares como texto puro, evitando posibles errores al manipular las cadenas.
2.	Manejo de Valores Nulos:
o	Se identificaron valores nulos en las columnas Top23, Top24 y Top25. Estos valores faltantes podrían haber causado problemas en cualquier análisis posterior, por lo que las filas con valores nulos fueron eliminadas.
o	La eliminación de estas filas redujo el número de registros de 1989 a 1986. Aunque este enfoque garantiza que todos los datos restantes estén completos, podría haber eliminado información potencialmente valiosa. En algunos casos, podríamos optar por otras técnicas, como rellenar valores nulos con datos estimados o medianas de la columna.
3.	Conversión de Fechas:
o	La columna Date, que originalmente estaba almacenada como una cadena de texto (object), fue convertida a un formato de fecha (datetime64). Esto es crucial para cualquier análisis de series temporales, ya que permite realizar cálculos y visualizaciones basadas en el tiempo con mayor precisión y facilidad.
4.	Eliminación de Duplicados:
o	Se verificó y eliminaron posibles duplicados en el dataset. Si bien en este caso no se encontraron duplicados, este es un paso importante en la limpieza de datos, ya que los duplicados pueden sesgar los resultados del análisis o del modelo predictivo.
5.	Resultado Final:
o	Después de la limpieza, el dataset quedó con 1986 filas y 27 columnas, todas las cuales ahora tienen valores completos y limpios.

```python
#Crear un gráfico de nube de palabras

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Definir las columnas 'Top' para el análisis
cols_to_analyze = ['Top1', 'Top2', 'Top3', 'Top4', 'Top5', 'Top6', 'Top7', 'Top8',
                   'Top16', 'Top17', 'Top18', 'Top19', 'Top20', 'Top21', 'Top22',
                   'Top23', 'Top24', 'Top25']

# Filtrar las filas donde el mercado sube (Label == 1)
df_up = df[df['Label'] == 1]

# Filtrar las filas donde el mercado baja (Label == 0)
df_down = df[df['Label'] == 0]

# Concatenar los textos de las columnas 'Top' en una sola cadena para cada subconjunto
text_data_up = ' '.join(df_up[col].dropna().astype(str).sum() for col in cols_to_analyze)
text_data_down = ' '.join(df_down[col].dropna().astype(str).sum() for col in cols_to_analyze)

# Crear la nube de palabras para el mercado sube
plt.figure(figsize=(10, 10))
plt.style.use('fast')
wc_up = WordCloud(background_color='green', width=1500, height=1500).generate(text_data_up)
plt.title('Palabras Más Nombradas en Noticias con Mercado Sube', fontsize=15)
plt.imshow(wc_up, interpolation='bilinear')
plt.axis('off')
plt.show()

# Crear la nube de palabras para el mercado baja
plt.figure(figsize=(10, 10))
plt.style.use('fast')
wc_down = WordCloud(background_color='red', width=1500, height=1500).generate(text_data_down)
plt.title('Palabras Más Nombradas en Noticias con Mercado Baja', fontsize=15)
plt.imshow(wc_down, interpolation='bilinear')
plt.axis('off')
plt.show()
```
